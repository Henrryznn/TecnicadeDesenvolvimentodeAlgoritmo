```python
'''Jogo da Bruxa (Adivinhe o Segredo da bruxa)'''

import random

def jogar_adivinhacao():
  """
  fumÃ§Ã£o principal do jogo de AdivinhaÃ§Ã£o com tema Hallowen.
  A bruxa guarda um nÃºmero secreto em seu caldeirÃ£o
  """

  print("ğŸ§¹ğŸ”®Bem vindo Ã  cabana da Bruxa!ğŸ§¹ğŸ”®")
  print("A Bruxa de Salem guardou um ++NÃºmero Secreto++ em seu caldeirÃ£o.")
  print("VocÃª tem poucas chances para descobrir, ou ++UM FEITIÃ‡O++ cairÃ¡ sobre vovÃª!\n")

  numero_secreto = random.randint(1,50)

  tentativas_maximas = 7
  tentativas_restantes = tentativas_maximas
  acertou = False

  while tentativas_restantes > 0 and not acertou:
    print(f"ğŸ‘»Tentativas restantes: {tentativas_restantes}")

    try:
      chute = int(input("Qual o seu palpite entre 1 e 50: "))
    except ValueError:
        print("ğŸš«Entrada invÃ¡lida!Digite apenas nÃºmeros inteiros.\n")
        continue

    if chute<1 or chute> 50:
      print("ğŸ’€O palpite deve estÃ¡ entre 1 e 50. Preste atenÃ§Ã£o!\n")
      continue

    tentativas_restantes -= 1

    if chute == numero_secreto:
      print(f"\nğŸƒPARABÃ‰NS! VocÃª quebrou o feitiÃ§o! O nÃºmero era *{numero_secreto}*.")
      acertou = True
    elif chute < numero_secreto:
      print("â¬‡ï¸ Muito Baixo! A fumaÃ§a do caldeirÃ£o indica que o nÃºmero Ã© *MAIOR*.")
    else:
      print("â¬†ï¸ Muito Alto! O caldeirÃ£o borbulha dizendo que o nÃºmero Ã© *MENOR*.")

  if not acertou:
    print("\nâš°ï¸ Suas chances acabaram! O feitiÃ§o estÃ¡ completo!")
    print(f"O NÃºmero Secreto da Bruxa era **{numero_secreto}**")

if __name__=="__main__":
  jogar_adivinhacao()
```

```python
'''Forca Assustadora (Jogo da Forca)'''

import random
import os
def exibir_forca(erros):
    """Desenha a forca de acordo com o nÃºmero de erros."""
    estagios = [
        # 0 erros
        """
           +---+
           |   |
               |
               |
               |
               |
        =========
        """,
        # 1 erro
        """
           +---+
           |   |
           O   |
               |
               |
               |
        =========
        """,
        # 2 erros
        """
           +---+
           |   |
           O   |
           |   |
               |
               |
        =========
        """,
        # 3 erros
        """
           +---+
           |   |
           O   |
          /|   |
               |
               |
        =========
        """,
        # 4 erros
        """
           +---+
           |   |
           O   |
          /|\\  |
               |
               |
        =========
        """,
        # 5 erros
        """
           +---+
           |   |
           O   |
          /|\\  |
          /    |
               |
        =========
        """,
        # 6 erros (GAME OVER)
        """
           +---+
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        =========
        """
    ]
    print(estagios[erros])


def jogar_forca():
    """FunÃ§Ã£o principal do Jogo da Forca com tema Halloween."""

    os.system('cls' if os.name == 'nt' else 'clear')

    print("ğŸ§™â€â™€ï¸  Bem-vindos Ã  Forca Assustadora! ğŸ•¯ï¸")
    print("Adivinhe o nome de um ser ou objeto de Halloween antes que o **FANTASMA** se complete!\n")

    palavras_halloween = ["ABOBORA", "VAMPIRO", "BRUXA", "ZUMBI", "CALDEIRAO", "MORCEGO", "FANTASMA", "CEMITERIO"]
    palavra_secreta = random.choice(palavras_halloween)

    letras_certas = []
    letras_erradas = []
    erros = 0
    max_erros = 6

    while erros < max_erros:
        exibir_forca(erros)

        palavra_atual = ""
        for letra in palavra_secreta:
            if letra in letras_certas:
                palavra_atual += letra + " "
            else:
                palavra_atual += "_ "

        print(f"\nğŸ•¸ï¸  Palavra: {palavra_atual}")
        print(f"âŒ Erros: {erros}/{max_erros}")
        print(f"ğŸ’€ Letras erradas: {', '.join(letras_erradas)}\n")

        if "_" not in palavra_atual:
            print(f"ğŸƒ VITÃ“RIA! A palavra era: **{palavra_secreta}**!!!")
            return

        tentativa = input("ğŸ”® Chute uma letra: ").strip().upper()

        if len(tentativa) != 1 or not tentativa.isalpha():
            print("âš ï¸  Por favor, chute apenas uma letra!\n")
            continue

        if tentativa in letras_certas or tentativa in letras_erradas:
            print(f"ğŸ‘» VocÃª jÃ¡ tentou a letra '{tentativa}'. Tente outra.\n")
            continue
        elif tentativa in palavra_secreta:
            print(f"âœ¨ Boa! A letra '{tentativa}' estÃ¡ na palavra.\n")
            letras_certas.append(tentativa)
        else:
            print(f"ğŸ’€ A letra '{tentativa}' nÃ£o estÃ¡ na palavra. O Fantasma avanÃ§a!\n")
            letras_erradas.append(tentativa)
            erros += 1

        print("-" * 30 + "\n")

    exibir_forca(erros)
    print("â˜ ï¸  FIM DE JOGO! O Fantasma te pegou!")
    print(f"A palavra secreta era: **{palavra_secreta}**.\n")


if __name__ == "__main__":
    jogar_forca()

```

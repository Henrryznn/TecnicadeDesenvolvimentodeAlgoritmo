```python
'''Jogo da Bruxa (Adivinhe o Segredo da bruxa)'''

import random

def jogar_adivinhacao():
  """
  fumção principal do jogo de Adivinhação com tema Hallowen.
  A bruxa guarda um número secreto em seu caldeirão
  """

  print("🧹🔮Bem vindo à cabana da Bruxa!🧹🔮")
  print("A Bruxa de Salem guardou um ++Número Secreto++ em seu caldeirão.")
  print("Você tem poucas chances para descobrir, ou ++UM FEITIÇO++ cairá sobre vovê!\n")

  numero_secreto = random.randint(1,50)

  tentativas_maximas = 7
  tentativas_restantes = tentativas_maximas
  acertou = False

  while tentativas_restantes > 0 and not acertou:
    print(f"👻Tentativas restantes: {tentativas_restantes}")

    try:
      chute = int(input("Qual o seu palpite entre 1 e 50: "))
    except ValueError:
        print("🚫Entrada inválida!Digite apenas números inteiros.\n")
        continue

    if chute<1 or chute> 50:
      print("💀O palpite deve está entre 1 e 50. Preste atenção!\n")
      continue

    tentativas_restantes -= 1

    if chute == numero_secreto:
      print(f"\n🎃PARABÉNS! Você quebrou o feitiço! O número era *{numero_secreto}*.")
      acertou = True
    elif chute < numero_secreto:
      print("⬇️ Muito Baixo! A fumaça do caldeirão indica que o número é *MAIOR*.")
    else:
      print("⬆️ Muito Alto! O caldeirão borbulha dizendo que o número é *MENOR*.")

  if not acertou:
    print("\n⚰️ Suas chances acabaram! O feitiço está completo!")
    print(f"O Número Secreto da Bruxa era **{numero_secreto}**")

if __name__=="__main__":
  jogar_adivinhacao()
```

```python
'''Forca Assustadora (Jogo da Forca)'''

import random
import os
def exibir_forca(erros):
    """Desenha a forca de acordo com o número de erros."""
    estagios = [
        # 0 erros
        """
           +---+
           |   |
               |
               |
               |
               |
        =========
        """,
        # 1 erro
        """
           +---+
           |   |
           O   |
               |
               |
               |
        =========
        """,
        # 2 erros
        """
           +---+
           |   |
           O   |
           |   |
               |
               |
        =========
        """,
        # 3 erros
        """
           +---+
           |   |
           O   |
          /|   |
               |
               |
        =========
        """,
        # 4 erros
        """
           +---+
           |   |
           O   |
          /|\\  |
               |
               |
        =========
        """,
        # 5 erros
        """
           +---+
           |   |
           O   |
          /|\\  |
          /    |
               |
        =========
        """,
        # 6 erros (GAME OVER)
        """
           +---+
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        =========
        """
    ]
    print(estagios[erros])


def jogar_forca():
    """Função principal do Jogo da Forca com tema Halloween."""

    os.system('cls' if os.name == 'nt' else 'clear')

    print("🧙‍♀️  Bem-vindos à Forca Assustadora! 🕯️")
    print("Adivinhe o nome de um ser ou objeto de Halloween antes que o **FANTASMA** se complete!\n")

    palavras_halloween = ["ABOBORA", "VAMPIRO", "BRUXA", "ZUMBI", "CALDEIRAO", "MORCEGO", "FANTASMA", "CEMITERIO"]
    palavra_secreta = random.choice(palavras_halloween)

    letras_certas = []
    letras_erradas = []
    erros = 0
    max_erros = 6

    while erros < max_erros:
        exibir_forca(erros)

        palavra_atual = ""
        for letra in palavra_secreta:
            if letra in letras_certas:
                palavra_atual += letra + " "
            else:
                palavra_atual += "_ "

        print(f"\n🕸️  Palavra: {palavra_atual}")
        print(f"❌ Erros: {erros}/{max_erros}")
        print(f"💀 Letras erradas: {', '.join(letras_erradas)}\n")

        if "_" not in palavra_atual:
            print(f"🎃 VITÓRIA! A palavra era: **{palavra_secreta}**!!!")
            return

        tentativa = input("🔮 Chute uma letra: ").strip().upper()

        if len(tentativa) != 1 or not tentativa.isalpha():
            print("⚠️  Por favor, chute apenas uma letra!\n")
            continue

        if tentativa in letras_certas or tentativa in letras_erradas:
            print(f"👻 Você já tentou a letra '{tentativa}'. Tente outra.\n")
            continue
        elif tentativa in palavra_secreta:
            print(f"✨ Boa! A letra '{tentativa}' está na palavra.\n")
            letras_certas.append(tentativa)
        else:
            print(f"💀 A letra '{tentativa}' não está na palavra. O Fantasma avança!\n")
            letras_erradas.append(tentativa)
            erros += 1

        print("-" * 30 + "\n")

    exibir_forca(erros)
    print("☠️  FIM DE JOGO! O Fantasma te pegou!")
    print(f"A palavra secreta era: **{palavra_secreta}**.\n")


if __name__ == "__main__":
    jogar_forca()

```

```python
provas = ["prova de lógica computacional", "prova de modelagem", "prova de circuitos lógicos"]
notas = []

while provas:
  print("\nProvas restantes: {provas}")
  prova_atual = provas[0]
  nota = float(input(f"Digite a nota da {prova_atual}:"))
  notas.append(nota)
  provas.pop(0)

print("\nTodas as provas foram realizadas!")
print("notas:", notas)

media = sum(notas)/len(notas)
print(f"Média final:{media:.2f}")

if media>=6:
  print("aprovado!")
else:
  print("reprovado.")
```
